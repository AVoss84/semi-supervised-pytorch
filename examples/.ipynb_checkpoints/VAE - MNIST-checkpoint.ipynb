# Variational AutoEncoder

This notebook shows how to use a Variational AutoEncoder to model the data distribution $p(x)$.

```python
import torch
import torch.nn.functional as F
from torchvision import datasets, transforms
cuda = torch.cuda.is_available()
```

```python
def transform(img):
    img = transforms.ToTensor()(img)
    img = img.view(-1)
    return img

mnist = datasets.MNIST('./', download=True, transform=transform)
unlabelled = torch.utils.data.DataLoader(mnist, batch_size=100, shuffle=True, num_workers=2)
labelled = next(iter(unlabelled))

mnist_val = datasets.MNIST('./', train=False, download=True, transform=transform)
validation = torch.utils.data.DataLoader(mnist_val, batch_size=100, shuffle=True, num_workers=2)
```

```python
from . import VariationalAutoencoder
from loss import VariationalInference, kl_divergence_normal

model = VariationalAutoencoder([28*28, 2, 128])

if cuda: model = model.cuda()

objective = VariationalInference(F.binary_cross_entropy, kl_divergence_normal)
optimizer = torch.optim.Adam(model.parameters(), lr=3e-5)
```

Train the VAE using the data.

```python
from torch.autograd import Variable

for epoch in range(50):
    for u, _ in unlabelled:
        if cuda: u = u.cuda()
        u = Variable(u)

        reconstruction, (z, z_mu, z_log_var) = model(u)
        loss = objective(reconstruction, u, z_mu, z_log_var)
        loss = torch.mean(loss)

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    if epoch % 10 == 0:
        l = loss.data[0]
        print("Epoch: {0:} loss: {1:.3f}".format(epoch, l))
```

```python
for epoch in range(100):    
    [x, y] = labelled

    x = Variable(x)
    y = Variable(y)

    if cuda:
        x = x.cuda()
        y = y.cuda()
        
    _, (z, _, _) = model(x)
    logits = classifier(z)
    loss = F.cross_entropy(logits, y)
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if epoch % 10 == 0:
        # Check validation accuracy
        accuracy = []
        for x, y in validation:
            _, (z, _, _) = model(Variable(x))
            logits = classifier(z)
            _, prediction = torch.max(classifier(z), 1)

            accuracy += [torch.mean((prediction.data == y).float())]

        print("Epoch: {0:} loss: {1:.3f}, accuracy: {2:.3f}".format(epoch, loss.data[0], np.mean(accuracy)))
```

Visualise the latent space (courtesy of [Fast forward labs VAE implementation](https://github.com/fastforwardlabs/vae-tf))

```python
from scipy.stats import norm

range_ = (0, 1)
min_, max_ = range_
dim = 28

# complex number steps act like np.linspace
# row, col indices (i, j) correspond to graph coords (y, x)
# rollaxis enables iteration over latent space 2-tuples
zs = np.rollaxis(np.mgrid[max_:min_:20*1j, min_:max_:20*1j], 0, 3)

DELTA = 1E-16 # delta to avoid +/- inf at 0, 1 boundaries
zs = np.array([norm.ppf(np.clip(z, DELTA, 1 - DELTA)) for z in zs])
zs = torch.from_numpy(zs).float()

canvas = np.vstack([np.hstack([x.reshape([dim, dim])
                               for x in model.generate(Variable(z_row)).data.numpy()])
                    for z_row in iter(zs)])

plt.figure(figsize=(20 / 2, 20 / 2))
# `extent` sets axis labels corresponding to latent space coords
plt.imshow(canvas, aspect="auto", extent=(range_ * 2))
ax = plt.gca()
ax.set_frame_on(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis("off")
plt.tight_layout()
```
